##**Theory**
**Unsupervised learning**

Unsupervised learning is a machine learning technique that involves training an algorithm on a dataset that does not have pre-labeled or pre-defined output classes or targets. In other words, the algorithm is presented with a set of input data and it learns to recognize patterns or relationships within the data without any guidance or supervision.

In unsupervised learning, the algorithm's goal is to discover hidden structures or patterns in the input data. This can be done through clustering, where the algorithm groups similar data points together, or through dimensionality reduction, where the algorithm identifies the most important features in the data and represents it in a lower-dimensional space.



**Clustering analysis**

Clustering is a technique used in unsupervised machine learning to group together similar data points into clusters or subgroups based on their inherent similarities. In clustering, the algorithm analyzes the input data and identifies patterns or similarities among the data points to group them into distinct clusters.

The primary goal of clustering is to find patterns or structures in the data that are not immediately apparent or to discover underlying subgroups in the data. Clustering algorithms do not require labeled data or predetermined classes, as they work solely on the input data itself.

Clustering can be used in a variety of applications such as customer segmentation, anomaly detection, image segmentation, and document clustering, among others. It is a powerful tool for data exploration and can provide valuable insights into the structure of large, complex datasets.

**DBSCAN (Density-Based Spatial Clustering of Applications with Noise)**

DBSCAN (Density-Based Spatial Clustering of Applications with Noise) is a popular density-based clustering algorithm that partitions a dataset into groups of points based on their density. The algorithm was proposed by Martin Ester, Hans-Peter Kriegel, JÃ¶rg Sander, and Xiaowei Xu in 1996.

DBSCAN defines clusters as areas of high density separated by areas of low density. Points in high-density regions are considered as "core points," while points in low-density regions that are close to core points are considered as "border points." Points that do not belong to any cluster are considered as "noise points."

The DBSCAN algorithm takes two parameters: "eps" and "min_samples." "eps" defines the radius of the neighborhood around a point, and "min_samples" specifies the minimum number of points that must be in the neighborhood for a point to be considered as a core point.

The algorithm works by starting with an arbitrary point in the dataset and expanding a cluster around it by adding nearby points to the cluster if they are also core points. Once there are no more core points to add, the algorithm selects a new unvisited point and repeats the process until all points have been visited.

The algorithm is used to identify dense regions of data points and separate them from less dense regions. Here are some general steps for performing a DBSCAN clustering analysis:

**Load the data:** Load the dataset you want to cluster.

**Preprocess the data:** Preprocess the data to remove any missing or invalid values.

**Choose the parameters:** Choose the parameters for the DBSCAN algorithm, including the epsilon value and minimum number of points.

**Compute the distance matrix:** Compute the distance matrix between the data points.

**Identify the core points:** Identify the core points in the dataset, which are points with at least the minimum number of neighbors within the epsilon distance.

**Identify the border points:** Identify the border points in the dataset, which are points that are within the epsilon distance of a core point but do not have enough neighbors to be considered core points.

**Identify the noise points:** Identify the noise points in the dataset, which are points that are not within the epsilon distance of any core or border points.

**Create the clusters:** Create the clusters by grouping together the core points and their neighboring border points.

**Analyze the clusters:** Analyze the clusters to understand their properties, such as their size, density, and shape.

**Visualize the clusters:** Visualize the clusters using graphs, plots, or other visualization techniques.

One of the main advantages of DBSCAN is its ability to identify clusters of arbitrary shape and handle noise in the dataset. However, the algorithm can be sensitive to the choice of parameters, and its performance can degrade when dealing with datasets of high dimensionality.

**Difference between K-means and DBSCAN algorithms**

K-means and DBSCAN (Density-Based Spatial Clustering of Applications with Noise) are both unsupervised machine learning algorithms used for clustering.

K-means is a centroid-based algorithm that seeks to partition the data into K clusters, where K is a predefined number. It works by iteratively assigning each data point to the nearest centroid and then recalculating the centroids based on the new cluster assignments. K-means is a popular clustering algorithm due to its simplicity and efficiency. However, it requires specifying the number of clusters beforehand, which can be a disadvantage in cases where the optimal number of clusters is not known.

In DBSCAN is a density-based algorithm that groups together data points that are closely packed together, while also identifying outliers as noise. The algorithm does not require a pre-specified number of clusters, making it more flexible than K-means. DBSCAN works by defining a radius around each data point and counting the number of points within that radius. If the number of points exceeds a pre-specified density threshold, then those points are considered part of a cluster. DBSCAN is especially useful in cases where the clusters have irregular shapes or varying densities.

The choice between K-means and DBSCAN depends on the specific problem and data characteristics.

##**Principal Component Analysis (PCA)**

Principal Component Analysis (PCA) is a statistical technique used to reduce the dimensionality of a dataset. It involves transforming a large set of variables into a smaller set of new variables called principal components (PCs), which are linear combinations of the original variables.

The main objective of PCA is to identify patterns in the data, as well as to identify the most important variables that contribute to the variation in the data. This is achieved by finding the directions in which the data varies the most, and projecting the data onto these directions, resulting in a new set of variables that capture most of the variance in the original data.

image.png

PCA can be used for a wide range of applications, such as data compression, data visualization, feature extraction, and data pre-processing for machine learning algorithms. It is particularly useful when dealing with high-dimensional data, where it is difficult to visualize and analyze the data directly, and where there may be redundant or irrelevant variables that can be eliminated to simplify the analysis.

**Find reasonable Epsilon values**

```python
outlier_percent = []
number_of_outliers = []

for eps in np.linspace(0.001,10,100):

    # Create Model
    dbscan = DBSCAN(eps=eps)
    dbscan.fit(two_blobs_outliers)

    # Number of Outliers
    number_of_outliers.append(np.sum(dbscan.labels_ == -1))

    # Percentage of points that are outliers
    perc_outliers = 100 * np.sum(dbscan.labels_ == -1) / len(dbscan.labels_)

    outlier_percent.append(perc_outliers)
```

```python
sns.lineplot(x=np.linspace(0.001,10,100),y=outlier_percent)
plt.ylabel("Percentage of Points Classified as Outliers")
plt.xlabel("Epsilon Value")
```

```python
sns.lineplot(x=np.linspace(0.001,10,100),y=number_of_outliers)
plt.ylabel("Number of Points Classified as Outliers")
plt.xlabel("Epsilon Value")
plt.xlim(0,1)
```

**"target" a percentage, like choose a range producing 1%-5% as outliers**

```python
sns.lineplot(x=np.linspace(0.001,10,100),y=outlier_percent)
plt.ylabel("Percentage of Points Classified as Outliers")
plt.xlabel("Epsilon Value")
plt.ylim(0,5)
plt.xlim(0,2)
plt.hlines(y=1,xmin=0,xmax=2,colors='red',ls='--')
```

**"target" a number of outliers, such as 3 points as outliers**

```python
sns.lineplot(x=np.linspace(0.001,10,100),y=number_of_outliers)
plt.ylabel("Number of Points Classified as Outliers")
plt.xlabel("Epsilon Value")
plt.ylim(0,10)
plt.xlim(0,6)
plt.hlines(y=3,xmin=0,xmax=10,colors='red',ls='--')
```

```python
pca_30 = PCA(n_components=30)
pca_30.fit(scaled_X)
```

```python
explained_variance = []

for n in range(1,30):
    pca = PCA(n_components=n)
    pca.fit(scaled_X)

    explained_variance.append(np.sum(pca.explained_variance_ratio_))
```

```python
plt.plot(range(1,30),explained_variance)
plt.xlabel("Number of Components")
plt.ylabel("Variance Explained");
```
